## Fixing Column Types/ Filling in Missing Data

```python
# Checking the datatypes of each column
df.dtypes
```

```python
# Mssub class, month sold,  overall qual and condition should be categorical types

df['ms_subclass'] = df['ms_subclass'].astype(str)
df['overall_cond'] = df['overall_cond'].astype(str)
df['overall_qual'] = df['overall_qual'].astype(str)
df['mo_sold'] = df['mo_sold'].astype(str
```

```python
# Creating a list of numerical features
numerical = df.select_dtypes(exclude='object')
numerical.columns
```

```python
categorical = df.select_dtypes('object')
categorical.columns
```

### Now that we have the correct column types we will fill them with None or 0's

```python
for col in categorical:
    df[col] = df[col].fillna('None')
for col in numerical:
    df[col] = df[col].fillna(0.0)
```

```python
# Checking to make sure all empty values are now filled
df.isnull().sum()
```
