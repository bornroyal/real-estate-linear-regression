## Handling Missing Data

## Basement


```python
bsmts = ames[['bsmtfin_type_1', 'bsmtfin_type_2', 'bsmtfin_sf_1', 'bsmtfin_sf_2', 'bsmt_unf_sf', 'bsmt_qual', 'bsmt_cond', 'bsmt_exposure', 'saleprice']]
```

```python
# Change basement condition to a numeric value so it can be encoded as polynomial feature
def bsmt_c_int(x):
    if x['bsmt_cond'] == 'Ex':
        return 5
    elif x['bsmt_cond'] == 'Gd':
        return 4
    elif x['bsmt_cond'] == 'TA':
        return 3
    elif x['bsmt_cond'] == 'Fa':
        return 2
    elif x['bsmt_cond'] == 'Po':
        return 1
    else:
        return 0
```

```python
# Change basement quality to a numeric value so it can be used for polynomial features
def bsmt_q_int(x):
    if x['bsmt_qual'] == 'Ex':
        return 5
    elif x['bsmt_qual'] == 'Gd':
        return 4
    elif x['bsmt_qual'] == 'TA':
        return 3
    elif x['bsmt_qual'] == 'Fa':
        return 2
    elif x['bsmt_qual'] == 'Po':
        return 1
    else:
        return 0
```

### Now we will create some polynomial features for basement

```python
ames['bsmt_qual_2'] = ames.apply(lambda x: bsmt_q_int(x), axis=1)
ames['bsmt_sf'] = ames['bsmtfin_sf_1'] + ames['bsmtfin_sf_2'] + ames['bsmt_unf_sf']
ames['bsmt_cond_2'] = ames.apply(lambda x: bsmt_c_int(x), axis=1)
ames['bsmt_sf_x_cond2'] = ames['bsmt_cond_2'] * ames['bsmt_sf']
ames['bsmt_sf_x_qual_2'] = ames['bsmt_qual_2'] * ames['bsmt_sf']
```
